```python
ch=[]
def get_characters(N):
    if (N>=221):
        return
    else:
        a=lines[N][0:]
        b=a.split(":")
        ch.append(b[0])
        get_characters(N+2)
    print(ch)
```

```python
Unique_Characters=set(ch)
```

```python
print(Unique_Characters)
print("Number of unique dialogue speakers in sample conversation:",len(Unique_Characters))
```

```python
Unique_Characters_list=list(Unique_Characters)
```

```python
def get(N):
    file=open("./"+N+".txt","w+")
    N=N+":"
    ch=""
    s1=""
    s2=""
    
    for i in lines:
            if (i.startswith(N)):
                ch=ch+" "+i

    space_found=N.find(" ")
    list_of_words=ch.split(" ")
    set_of_words=set(list_of_words)
    set_of_words.discard(N)
    set_of_words.discard('')    

    if(space_found>=0):
        s1=s1+N[0:space_found]
        s2=s2+N[space_found+1:]
        set_of_words.discard(s1)
        set_of_words.discard(s2)
                    
    print("Character",N)
    for words in set_of_words:
        print (words)
        words=words.replace("?","")
        words=words.replace(".","")
        words=words.replace(",","")
        words=words.replace("!","")
        
        file.writelines(words+"\n")
    file.close()
    
for x in Unique_Characters_list:
    get(x)
```
